```python
source = urllib.request.urlopen(url).read()
soup = bs.BeautifulSoup(source, 'html')
```

```python
table = soup.table
table_rows = table.find_all('tr')
metar_list = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    metar_list.append(row)
```

```python
metar_db = []
for idx, line in enumerate(metar_list):
    #print(idx, line)
    for item in line:
        found = item.find('METAR '+str(ride_airport))
        if idx == 3: testline = line
        if found!=-1:
            #METAR CYXU 042200Z 09006KT 040V120 15SM FEW210 BKN250 04/M08
            try:
                print('FOUND!',found)
                print(item)
                mday = int(item[found+11:found+13])
                mhour = int(item[found+13:found+15])
                mminute = int(item[found+15:found+17])
                wind_dir = int(item[found+19:found+22])
                wind_spd = int(item[found+22:found+24])
                print('mday:', mday)
                print('mhour:', mhour)
                print('mminute:', mminute)
                print('wind_dir:', wind_dir)
                print('wind_spd:', wind_spd)
                metar_db.append([wind_dir, wind_spd, datetime.datetime(year=2019, month=4, day=mday, hour=mhour, minute=mminute, second=0, tzinfo=pytz.UTC)])
            except:
                print('parsing error')
```
